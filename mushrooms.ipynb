# Mushroom foraging with Hypertools

### This dataset contains descriptive features of various mushroom species, as well as labels indicating whether tor not they are poisonous. It was downloaded from https://www.kaggle.com/uciml/mushroom-classification

### Import the libraries we need

```python
import pandas as pd
import hypertools as hyp
%matplotlib inline
```

### Read in the data with pandas

```python
data = pd.read_csv('data/mushrooms.csv')
data.head()
```

### In order to visualize this data, we first will need to convert the features into numerical values.  To do this, we will use pandas dummy variable function, which turns a columns of text labels into a set of columns, where each new column represents a different label within a column.

```python
class_labels = data.pop('class')
```

### Now let's plot the high-dimensional data in a low dimensional space by converting it to a numpy array and passing it to hyperplot.

```python
hyp.plot(data,'o', normalize=False) # if the number of features is greater than 3, the default is to plot in 3d
```

### From the plots above, it's clear that there are multiple clusters in this data. In addition to descriptive features, this dataset also contains whether or not the mushroom is poisonous.  Do these clusters have bearing on whether or not the mushroom is poisonous?  Let's try to understand this by coloring the points based on whether the mushroom is poisonous:

```python
hyp.plot(data,'o', group=class_labels, normalize=False, legend=list(set(class_labels)))
```

###  From the plot above it is clear that the clustering of the mushrooms carries information about whether or not they are poisonous (red refers to non-poisonous and blue is poisonous). It also looks like there are a number of distinct clusters that are poisonous/non-poisonous (perhaps ~10-15?).  

### Let's use the 'cluster' feature of hypertools to discover these clusters using a k-means fitting procedure:

```python
hyp.plot(data, 'o', n_clusters=15, normalize=False)

# you can also recover the cluster labels using the cluster tool
cluster_labels = hyp.tools.cluster(data, n_clusters=15, ndims=3) 

# hyp.plot(data, 'o', point_colors=cluster_labels, ndims=2)
```

### Sidenote: we can change the color palette using the palette argument.  Hypertools supports matplotlib and seaborn color palettes.

```python
hyp.plot(data,'o', group=cluster_labels, palette="deep", normalize=False)
```

### Hypertools uses PCA to reduce the dimensionality by default, but there are other ways to do dimensionality reduction.  Let's try reducing with various techniques, but keeping the cluster labels the same.

```python
from sklearn.decomposition import FastICA
ICA_model = FastICA(n_components=3)
reduced_data_ICA = ICA_model.fit_transform(hyp.tools.df2mat(data))
```

## ICA

```python
hyp.plot(reduced_data_ICA,'o', group=class_labels, legend=list(set(class_labels)), normalize=False)
```

```python
from sklearn.manifold import TSNE
TSNE_model = TSNE(n_components=3)
reduced_data_TSNE = TSNE_model.fit_transform(hyp.tools.df2mat(data))
```

## t-SNE

```python
hyp.plot(reduced_data_TSNE,'o', group=class_labels, legend=list(set(class_labels)), normalize=False)
```

```python
#Warning: this can take a long time to run
from sklearn.manifold import MDS
MDS_model = MDS(n_components=3, n_jobs=-1)
reduced_data_MDS = MDS_model.fit_transform(hyp.tools.df2mat(data))
```

### MDS

```python
hyp.plot(reduced_data_MDS,'o', point_colors = class_labels, legend=list(set(class_labels)), normalize=False)
```
